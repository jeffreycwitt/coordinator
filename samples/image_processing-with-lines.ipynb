```python
im = imread( 'rothwell1r.jpg')
```

```python
plt.imshow( im )
```

```python
imgplot = plt.hist(im.ravel(), bins=256, range=(0.0, 256), fc='k', ec='k')
```

```python
meanrows = im[:,:,0].mean(axis=1)
```

```python
plt.plot(meanrows)
```

```python
for mean in meanrows:
    print(mean)
```

```python
newdata = []
```

```python
for i in range(1,len(meanrows)):
    if meanrows[i] > meanrows[i-1] and meanrows[i] > meanrows[i+1]:
        newdata.append(i)
```

```python
len(newdata)
```

```python
for i in newdata:
    print(i)
    im[i,:,:] = 0
```

```python
for i in range(0,len(im)):
    print(im[i,i,:])
```
